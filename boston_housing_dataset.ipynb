```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Loading the dataset

```python
data_url = "http://lib.stat.cmu.edu/datasets/boston"
raw_df = pd.read_csv(data_url, sep="\s+", skiprows=22, header=None)
data = np.hstack([raw_df.values[::2, :], raw_df.values[1::2, :2]])
target = raw_df.values[1::2, 2]
```

feature names from: https://www.cs.toronto.edu/~delve/data/boston/bostonDetail.html

```python
feature_names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"]

boston_df = pd.DataFrame(data = data, columns = feature_names)
boston_df["MEDV"] = target
```

```python
boston_df
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.2)
```

```python
print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

Using Linear Regression

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()
linear_regressor.fit(X_train_scaled, y_train)
predictions = linear_regressor.predict(X_test_scaled)
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, predictions)
print("The accuracy is {}%".format(round(score, 2) *100))
```
